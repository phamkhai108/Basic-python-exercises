# Thực hành 1


```python
import numpy as np

# Nhập cấp của ma trận vuông
n = int(input("Nhập cấp của ma trận vuông: "))

# Tạo ma trận ngẫu nhiên
matrix = np.random.randint(0, 101, size=(n, n))

# In ma trận
print("Ma trận:")
print(matrix)
#Kích thước của ma trận
print("kích thước của ma trận là:", matrix.shape) 
# In kiểu dữ liệu của phần tử
print("Kiểu dữ liệu của phần tử là:", matrix.dtype)
#In ra số phần tử trong ma trận
print("Số phần tử trong ma trận:",matrix.size)
#In ra số chiều của ma trận
print("Số chiều của ma trận:", matrix.ndim)
```

```python
print("vector nằm trên đường chéo chính:\n",matrix.diagonal())
print("vector nằm trên đường chéo phụ:\n",np.diag(np.flipud(matrix)))
```

```python
x = int(input("nhập vào giá trị x (0-100):" ))
print(f"số phần tử lớn bằng giá trị {x} là:",np.sum(matrix == x))
print(f"số phần tử nhỏ hơn giá trị {x} là:",np.sum(matrix < x))
print(f"số phần tử lớn hơn giá trị {x} là:",np.sum(matrix > x))
```

# Thực hành 2


```python
import numpy as np
#đọc file
path = 'Diem_2A.txt'
diem_2a = np.loadtxt(path, delimiter=',',dtype=np.int8)
#in ma trận
print(diem_2a)
#điểm trung bình 
dtb = np.mean(diem_2a, axis=0)
print("Điểm trung bình của từng học sinh trong lớp: \n", dtb, "\n-------------------------------------------------------------")
#điểm trung bình học sinh cao nhất
print("điểm trung bình cao nhất:", dtb.max())
index_max = dtb.argmax()
print("Của học sinh thứ:", index_max+1)
print("Bảng điểm đầy đủ của họ sinh:", diem_2a[:,index_max], "\n-------------------------------------------------------------")
#điểm trung bình hs thấp nhát
index_min = dtb.argmin()
print("điểm trung bình thấp nhất:", dtb.min())
print("Của học sinh thứ:", index_min+1)
print("Bảng điểm đầy đủ của học sinh:",diem_2a[:,index_min])
```

```python
dtb = np.mean(diem_2a, axis=1)
print("Điểm trung bình của từng môn học: \n", dtb, "\n-------------------------------------------------------------")
#điểm trung bình học sinh cao nhất
print("điểm trung bình môn cao nhất:", dtb.max())
index_max = dtb.argmax()
print("Của môn học thứ:", index_max+1)
print("Bảng điểm đầy đủ môn học:", diem_2a[index_max,:], "\n-------------------------------------------------------------")
#điểm trung bình hs thấp nhát
index_min = dtb.argmin()
print("điểm trung bình môn thấp nhất:", dtb.min())
print("Của môn học thứ:", index_min+1)
print("Bảng điểm đầy đủ của môn học:",diem_2a[index_min,:])
```

```python
#Sinh viên có điểm học đồng đều nhất và bảng điểm hs đó
indexStd_min_hs = np.argmin(np.std(diem_2a,axis = 0))
print("Học sinh học đồng đều nhất ở vị trí:",indexStd_min_hs+1)
print("Bảng điểm của học sinh học đồng đều nhất:",diem_2a[:,indexStd_min_hs],'\n')
#Sinh viên có điểm học lệch nhất và bảng điểm hs đó
indexStd_max_hs = np.argmax(np.std(diem_2a,axis = 0))
print("Học sinh học lệch nhất ở vị trí:",indexStd_max_hs+1)
print("Bảng điểm của học sinh học lệch nhất:",diem_2a[:,indexStd_max_hs])

print('-----------------------------------------------')


#môn học có điểm đồng đều nhất và bảng điểm môn đó
indexStd_min_mh = np.argmin(np.std(diem_2a,axis =1))
print("Môn học điểm đồng đều nhất ở vị trí:",indexStd_min_mh+1)
print("Bảng điểm của môn học có điểm đồng đều nhất:",diem_2a[indexStd_min_mh,:],'\n')
#Môn học có điểm lệch nhất và bảng điểm môn đó
indexStd_max_mh = np.argmax(np.std(diem_2a,axis = 1))
print("Môn học có điểm lệch nhất ở vị trí:",indexStd_max_mh+1)
print("Bảng điểm của môn học có điểm lệch nhất:",diem_2a[indexStd_max_mh,:])
```

```python
import numpy as np

# Tạo ma trận 3x7 với giá trị rỗng (0)
data_thongke = np.zeros((3, 7))
# gans giá trị cho hàng đầu tiên
data_thongke[0,:-1] = np.max(Temp,axis=0)
data_thongke[0, -1] = Temp.max()

#gán giá trị cho hàng thứ 2
data_thongke[1,:-1] = np.round(np.mean(Temp,axis=0),2)
data_thongke[1, -1] = np.round(Temp.mean(),2)

#gán giá trị cho hàng thứ 3
data_thongke[2,:-1] = np.min(Temp,axis=0)
data_thongke[2, -1] = Temp.min()

print(data_thongke)
print(type(data_thongke))
print('Kích thước:',data_thongke.shape)

#lưu vào file thongke.txt
np.savetxt('thongke.txt', data_thongke, fmt='%.2f', delimiter='\t')

#in ra dữ liệu trong filel thongke.txt để kiểm tra
with open('thongke.txt', 'r') as file:
    content = file.read()
print(''content)
```
