```python
import numpy as np

# Nhp cp ca ma trn vung
n = int(input("Nhp cp ca ma trn vung: "))

# To ma trn ngu nhin
matrix = np.random.randint(0, 101, size=(n, n))

# In ma trn
print("Ma trn:")
print(matrix)
#Kch thc ca ma trn
print("kch thc ca ma trn l:", matrix.shape) 
# In kiu d liu ca phn t
print("Kiu d liu ca phn t l:", matrix.dtype)
#In ra s phn t trong ma trn
print("S phn t trong ma trn:",matrix.size)
#In ra s chiu ca ma trn
print("S chiu ca ma trn:", matrix.ndim)
```

```python
print("vector nm trn ng cho chnh:\n",matrix.diagonal())
print("vector nm trn ng cho ph:\n",np.diag(np.flipud(matrix)))
```

```python
x = int(input("nhp vo gi tr x (0-100):" ))
print(f"s phn t ln bng gi tr {x} l:",np.sum(matrix == x))
print(f"s phn t nh hn gi tr {x} l:",np.sum(matrix < x))
print(f"s phn t ln hn gi tr {x} l:",np.sum(matrix > x))
```

# Thc hnh 2


```python
import numpy as np
#c file
path = 'Diem_2A.txt'
diem_2a = np.loadtxt(path, delimiter=',',dtype=np.int8)
#in ma trn
print(diem_2a)
#im trung bnh 
dtb = np.mean(diem_2a, axis=0)
print("im trung bnh ca tng hc sinh trong lp: \n", dtb, "\n-------------------------------------------------------------")
#im trung bnh hc sinh cao nht
print("im trung bnh cao nht:", dtb.max())
index_max = dtb.argmax()
print("Ca hc sinh th:", index_max+1)
print("Bng im y  ca h sinh:", diem_2a[:,index_max], "\n-------------------------------------------------------------")
#im trung bnh hs thp nht
index_min = dtb.argmin()
print("im trung bnh thp nht:", dtb.min())
print("Ca hc sinh th:", index_min+1)
print("Bng im y  ca hc sinh:",diem_2a[:,index_min])
```

```python
dtb = np.mean(diem_2a, axis=1)
print("im trung bnh ca tng mn hc: \n", dtb, "\n-------------------------------------------------------------")
#im trung bnh hc sinh cao nht
print("im trung bnh mn cao nht:", dtb.max())
index_max = dtb.argmax()
print("Ca mn hc th:", index_max+1)
print("Bng im y  mn hc:", diem_2a[index_max,:], "\n-------------------------------------------------------------")
#im trung bnh hs thp nht
index_min = dtb.argmin()
print("im trung bnh mn thp nht:", dtb.min())
print("Ca mn hc th:", index_min+1)
print("Bng im y  ca mn hc:",diem_2a[index_min,:])
```

```python
#Sinh vin c im hc ng u nht v bng im hs 
indexStd_min_hs = np.argmin(np.std(diem_2a,axis = 0))
print("Hc sinh hc ng u nht  v tr:",indexStd_min_hs+1)
print("Bng im ca hc sinh hc ng u nht:",diem_2a[:,indexStd_min_hs],'\n')
#Sinh vin c im hc lch nht v bng im hs 
indexStd_max_hs = np.argmax(np.std(diem_2a,axis = 0))
print("Hc sinh hc lch nht  v tr:",indexStd_max_hs+1)
print("Bng im ca hc sinh hc lch nht:",diem_2a[:,indexStd_max_hs])

print('-----------------------------------------------')


#mn hc c im ng u nht v bng im mn 
indexStd_min_mh = np.argmin(np.std(diem_2a,axis =1))
print("Mn hc im ng u nht  v tr:",indexStd_min_mh+1)
print("Bng im ca mn hc c im ng u nht:",diem_2a[indexStd_min_mh,:],'\n')
#Mn hc c im lch nht v bng im mn 
indexStd_max_mh = np.argmax(np.std(diem_2a,axis = 1))
print("Mn hc c im lch nht  v tr:",indexStd_max_mh+1)
print("Bng im ca mn hc c im lch nht:",diem_2a[indexStd_max_mh,:])
```
